```python
import torch
import pandas as pd
import statsmodels.api as sm
import pylab
import numpy as np
size=16
seed=31
Sigma=1
mu=0
```

```python
def plot(U1,U2):
  X= np.sqrt(-2 * np.log(U1))*np.cos(2 * np.pi * U2)
  X=X*Sigma+mu
  X_hist = pd.Series(X)
  X_hist.hist()
```

```python
def qqplot(U1,U2):
  X= np.sqrt(-2 * np.log(U1))*np.cos(2 * np.pi * U2)
  X=X*Sigma+mu
  X_qplot = np.array(X)
  sm.qqplot(X_qplot, line='45')
  pylab.show()
```

**GPU RAND**

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
torch.manual_seed(seed)
U1=torch.rand(size).to(device)
U2=torch.rand(size).to(device)
plot(U1.to("cpu").numpy(),U2.to("cpu").numpy())
qqplot(U1.to("cpu").numpy(),U2.to("cpu").numpy())
```

**CPU RAND**

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
torch.manual_seed(seed)
U1=torch.rand(size).numpy()
U2=torch.rand(size).numpy()
plot(U1,U2)
qqplot(U1,U2)
```
